```python
import pandas as pd

# Notebook para gerar arquivo txt para treinar o llama3. Essa foi a forma que consegui desenvolver para nao consumir api da groq.
# Com o txt o assistente consegue entender melhor sobre o grafico sem precisar consumuir muito
# Caso o projeto fosse para uso comercial, o arquivo csv seria carregado diretamente no LLM || Esse notebook e o txt nao seriam mais necessarios

# Tratamento dos dados:
df = pd.read_csv('UnileverNASDAQ.csv')

df['Date'] = pd.to_datetime(df['Date'])

dataMaior = df['Date'].max()
dataMenor = df['Date'].min()	

df['Close/Last'] = df['Close/Last'].replace({'\$': '', ',': ''}, regex=True)

df['Close/Last'] = df['Close/Last'].astype(float)  

df['Open'] = df['Open'].replace({'\$': '', ',': ''}, regex=True)

df['Open'] = df['Open'].astype(float)  

df['High'] = df['High'].replace({'\$': '', ',': ''}, regex=True)

df['High'] = df['High'].astype(float)  

df['Low'] = df['Low'].replace({'\$': '', ',': ''}, regex=True)

df['Low'] = df['Low'].astype(float)  

# Volume:

# Total volume:
totalVolume = df['Volume'].sum()

# Total volume formatado:
totalVolumef = f"${totalVolume:,.2f}"

# Media do volume:
mediaVolume = df['Volume'].mean()

# Media formatada:
mediaVolumef = f"${mediaVolume:,.2f}"

# media volume ano
df['Year'] = df['Date'].dt.year

mediaVolumeAno = df.groupby('Year')['Volume'].mean()

# Formtado:
mediaVolumeAnof = mediaVolumeAno.apply(lambda x: f"${x:,.2f}")

# Volume max e min e seu ano:
volumeMaximo = df['Volume'].max()
volumeMaximof = f"${volumeMaximo:,.2f}"
volumeMaximoAno = df.loc[df['Volume'].idxmax(), 'Date'].year

volumeMinimo = df['Volume'].min()
volumeMinimoAno = df.loc[df['Volume'].idxmin(), 'Date'].year
volumeMinimof = f"${volumeMinimo:,.2f}"
```

```python
# Gerando o csv
caminho_arquivo = 'treinamentoBOT.txt'

with open(caminho_arquivo, 'w') as file:
    file.write(f"Data maior: {dataMaior}\n")
    file.write(f"Data menor: {dataMenor}\n")
    
    file.write(f"Total volume: {totalVolumef}\n")
    file.write(f"Media do volume: {mediaVolumef}\n")
    file.write(f"Media do volume por ano:\n{mediaVolumeAnof}\n")
    
    file.write(f"Volume maximo: {volumeMaximof}\n")
    file.write(f"Volume minimo: {volumeMinimof}\n")
    file.write(f"Ano do volume maximo: {volumeMaximoAno}\n")
    file.write(f"Ano do volume minimo: {volumeMinimoAno}\n")
```

```python
print(f"Data maior: {dataMaior}\nData menor: {dataMenor}\n")

print(f"Total volume: {totalVolumef}\n")

print(f"Média do volume: {mediaVolumef}\n")

print(f"Média do volume por ano:\n{mediaVolumeAnof}\n")

print(f"Volume máximo: {volumeMaximof}\nVolume mínimo: {volumeMinimof}")
print(f"Ano do volume máximo: {volumeMaximoAno}\nAno do volume mínimo: {volumeMinimoAno}\n")
```
